```python
# Machine learning liberary
# function for splitting dataset and cross validation score on training set.
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score # Find prediction on training set.
from sklearn.model_selection import StratifiedKFold # create no of folds or splits.

# functions for ML algorithm implementation
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

# Step 1 - Reading and Understanding the Dataset

names = ['sepal-length','sepal-width','petal-length','petal-width','class']
ds = pd.read_csv(r'/content/iris.csv',names=names)


# Step 5 - Supervised Machine Learning.
# split the data into two parts 1.First 4 variables 2. class variables
array = ds.values # Converting a dataset into array
print(array)
x = array[:,0:4] # input variable sl(0),sw(1),pl(2),pw(3)
y = array[:,4] # output variable class
print(x)
print(y)
```

```python
# x- first 4 columns(input variable) with 600 records
# y- last column or class with all 600 records

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.20, random_state=1)


# X_train - first 4 column 80%(SL,SW,PL,PW) - 480 lines
# X_test - first  4 column 20% - 120 lines
# Y_train - last column 80% (class) - 480 lines
# Y_test - last column 20% - 120 lines

model = DecisionTreeClassifier() # SL Algorithm
kfold = StratifiedKFold(n_splits=10, random_state=1,shuffle=True)
cv_results = cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')
print(cv_results) # 10
print('Decision Tree accuracy=',cv_results.mean()*100)

print()

model = KNeighborsClassifier() # SL Algorithm
kfold = StratifiedKFold(n_splits=10, random_state=1,shuffle=True)
cv_results = cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')
print(cv_results) # 10
print('KNN=',cv_results.mean()*100)

print()

model = GaussianNB() # SL Algorithm
kfold = StratifiedKFold(n_splits=10, random_state=1,shuffle=True)
cv_results = cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')
print(cv_results) # 10
print('Naive Bayes Accuracy=',cv_results.mean()*100)


```

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# machine learning library
# functions for splitting dataset and cross validation scores on training set
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score  # find prediciton on training set
from sklearn.model_selection import StratifiedKFold   # create no of folds or splits

# functions to generate the output 
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

# functions for algorithms implementation
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

# Step 1 - Undrestanding the Dataset
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
ds = pd.read_csv(r'/content/iris.csv',names=names)


# step 5 - Supervised Machine Learning
# Split the data into two parts 1. 4 variables 2.class variable
array = ds.values # converting the entire dataset into array
X = array[:,0:4] # input variable sl(0),sw(1),pl(2),pw(3)()
y = array[:,4] # output variable class
print(X)
print(y)

# x = first 4 column with 600 records
# y = last column or class with all 600 records

X_train, X_test, Y_train, Y_test = train_test_split(X ,y, test_size=0.20,random_state=1)

# X_train - first 4 column 80%(SL,SW,PL,PW) - 480 lines
# X_test - first  4 column 20% - 120 lines
# Y_train - last column 80% (class) - 480 lines
# Y_test - last column 20% - 120 lines
models = []
models.append(('LR', LogisticRegression(solver='liblinear',multi_class='ovr')))
models.append(('KNN', KNeighborsClassifier()))
models.append(('DT', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC(gamma='auto')))


results = []
names = []

for name,model in models:
    kfold = StratifiedKFold(n_splits=10,random_state=1,shuffle=True)
    cv_results = cross_val_score(model,X_train,Y_train,cv=kfold,scoring='accuracy')
    results.append(cv_results)
    names.append(name)
              
    print('%s: %f (%f)' %(names,cv_results.mean(),cv_results.std()))  



    
```
